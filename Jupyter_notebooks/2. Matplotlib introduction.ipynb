## Подключаем стандартные библиотеки

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Создаём равномерно распределённые точки на прямой

```python
x = np.linspace(-5,5,100) # отрезок [-5, 5] разделяется на 100 равноотстоящих точек
print(x)
```

```python
y = np.sin(x) # применяем функцию f(x) = sin(x) к исходному массиву поэлементно
print(y)
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y,s=2,color='red')
```

```python
plt.figure(figsize=(12,8)) # Регулируем размер исходного графика
plt.title('Пример использования библиотеки\nMatplotlib', size=18) # Указываем название графика
plt.grid(color='blue', linestyle='--', linewidth=0.5) # Отрисовываем сетку
plt.scatter(x, np.sin(x), s=10, color='green',label='f(x)=sin(x)') # Первый график
plt.scatter(x, np.cos(x), s=38, color='red', marker ='*', label='g(x)=cos(x)') # Второй график
plt.legend(loc='best', prop={'size': 16}) # Создаем легенду (в самом свободном месте)

plt.savefig('/Users/gordon/LNMO/test_pic.png') # Сохраняем график как .png картинку
```
