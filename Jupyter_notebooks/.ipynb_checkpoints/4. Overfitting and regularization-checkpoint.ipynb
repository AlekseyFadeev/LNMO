```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression



# Истинная функция, которую будем моделировать
def true_fun(X): 
    return np.cos(1.5 * np.pi * X)

np.random.seed(0)

n_samples = 20
degrees = [1, 4, 10, 12]

X = np.sort(np.random.rand(n_samples))
y = true_fun(X) + np.random.randn(n_samples) * 0.1

# Рисуем график для каждого значения степени
plt.figure(figsize=(14, 5))
for i in range(len(degrees)):
    ax = plt.subplot(1, len(degrees), i + 1)

    polynomial_features = PolynomialFeatures(degree=degrees[i],
                                             include_bias=False)
    linear_regression = LinearRegression()
    
    # Объединяем операции выделения полиномиальных признаков и обучение модели в один pipeline:
    pipeline = Pipeline([("polynomial_features", polynomial_features),
                         ("linear_regression", linear_regression)])
    pipeline.fit(X[:, np.newaxis], y) # альтернатива X.reshape(-1,1)
    
    print("Degree: ", degrees[i], " Coefficients: ", pipeline.steps[1][1].coef_)

    # Evaluate the models using crossvalidation
    scores = cross_val_score(pipeline, X[:, np.newaxis], y,
                             scoring="neg_mean_squared_error", cv=10)

    X_test = np.linspace(0, 1, 100)
    plt.plot(X_test, pipeline.predict(X_test[:, np.newaxis]), label="Model", linewidth=2)
    plt.plot(X_test, true_fun(X_test), label="True function", linewidth=2)
    plt.scatter(X, y, edgecolor='b', s=30, label="Samples")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.xlim((0, 1))
    plt.ylim((-2, 2))
    plt.legend(loc="best")
    plt.title("Degree {}\nMSE = {:.2e}(+/- {:.2e})".format(
        degrees[i], -scores.mean(), scores.std()))
```

## Таким образом, чем больше степень признаков в модели (<font color=blue>polynomial features</font>), тем вероятнее <font color=red>переобучение</font>.
Наблюдается т.н. **эффект Рунге**: коэффициенты многочленов очень быстро растут (по модулю). Попробуем справится с этим эффектом с помощью техники **регуляризации** (<font color=blue>регуляризация Тихонова) </font>:<br>
\begin{equation}
Loss_{2} = Loss_{old} + \lambda \cdot ||\omega||_{2}^{2}\\
где \,\,||\omega||_{2} = \sqrt{\sum_{i=1}^{n}\omega_{i}^{2}}
\end{equation}

Коэффициент $\lambda$ показывает насколько важнее для нас минимизировать нормы коэффициентов по сравнению с изначальной функцией потерь (**trade-off**). Модель линейной регрессии, в которой ищется вектор коэффициентов $\omega$, минимизирующий $Loss_{2}$, называется <font color=blue> гребневой (ridge) регрессией </font>.<br>


```python
from sklearn.linear_model import Ridge

plt.figure(figsize=(14, 5))
for i in range(len(degrees)):
    ax = plt.subplot(1, len(degrees), i + 1)

    polynomial_features = PolynomialFeatures(degree=degrees[i],
                                             include_bias=False)
    
    # *****************************************************
    # Единственная строчка, которую мы поменяли
    ridge_regression = Ridge(alpha=0.005)
    # *****************************************************
    
    
    # Объединяем операции выделения полиномиальных признаков и обучение модели в один pipeline:
    pipeline = Pipeline([("polynomial_features", polynomial_features),
                         ("linear_regression", ridge_regression)])
    pipeline.fit(X[:, np.newaxis], y)
    
    print("Degree: ", degrees[i], " Coefficients: ", pipeline.steps[1][1].coef_)

    # Evaluate the models using crossvalidation
    scores = cross_val_score(pipeline, X[:, np.newaxis], y,
                             scoring="neg_mean_squared_error", cv=10)

    X_test = np.linspace(0, 1, 100)
    plt.plot(X_test, pipeline.predict(X_test[:, np.newaxis]), label="Model", linewidth=2)
    plt.plot(X_test, true_fun(X_test), label="True function", linewidth=2)
    plt.scatter(X, y, edgecolor='b', s=30, label="Samples")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.xlim((0, 1))
    plt.ylim((-2, 2))
    plt.legend(loc="best")
    plt.title("Degree {}\nMSE = {:.2e}(+/- {:.2e})".format(
        degrees[i], -scores.mean(), scores.std()))
```

## Обратите внимание на небольшие коэффициенты и устойчивость к переобучению.

На практике рассматривают ещё такую модификацию функции потерь:
\begin{equation}
Loss_{1} = Loss_{old} + \lambda \cdot ||\omega||_{1}^{2}\\
где \,\,||\omega||_{1} = \sqrt{\sum_{i=1}^{n}|\omega_{i}|}
\end{equation}
Регрессия с такой функцией потерь называется <font color=blue> лассо-регрессией</font> и обладает тем свойством, что обнуляется большое количество "несущественных" признаков.
Реализуем эту модель, изменив всего одну строчку в предыдущем коде.

```python
from sklearn.linear_model import Lasso

plt.figure(figsize=(14, 5))
for i in range(len(degrees)):
    ax = plt.subplot(1, len(degrees), i + 1)

    polynomial_features = PolynomialFeatures(degree=degrees[i],
                                             include_bias=False)
    
    # *****************************************************
    # Единственная строчка, которую мы поменяли
    lasso_regression = Lasso(alpha=0.005)
    # *****************************************************
    
    
    # Объединяем операции выделения полиномиальных признаков и обучение модели в один pipeline:
    pipeline = Pipeline([("polynomial_features", polynomial_features),
                         ("linear_regression", lasso_regression)])
    pipeline.fit(X[:, np.newaxis], y)
    
    print("Degree: ", degrees[i], " Coefficients: ", pipeline.steps[1][1].coef_)

    # Evaluate the models using crossvalidation
    scores = cross_val_score(pipeline, X[:, np.newaxis], y,
                             scoring="neg_mean_squared_error", cv=10)

    X_test = np.linspace(0, 1, 100)
    plt.plot(X_test, pipeline.predict(X_test[:, np.newaxis]), label="Model", linewidth=2)
    plt.plot(X_test, true_fun(X_test), label="True function", linewidth=2)
    plt.scatter(X, y, edgecolor='b', s=30, label="Samples")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.xlim((0, 1))
    plt.ylim((-2, 2))
    plt.legend(loc="best")
    plt.title("Degree {}\nMSE = {:.2e}(+/- {:.2e})".format(
        degrees[i], -scores.mean(), scores.std()))
```

## Получили большое количество признаков с 0 коэффициентом при схожем общем качестве приближения.
Вот схематичный рисунок, поясняющий эффект обнуления коэффициентов в лассо-регрессии:


Также иногда встречается комбинация двух описанных ранее моделей (<font color=blue> Elastic net </font>):
\begin{equation}
Loss_{elastic} = Loss_{old} + \lambda_{1}\cdot||\omega||_{1}^{2} + \lambda_{2}\cdot||\omega||_{2}^{2}
\end{equation}
Имеет смысл использовать в ситуациях большого количества признаков и низкой размерности исходных данных (т.к. модель лассо обнулила бы много важных признаков).
